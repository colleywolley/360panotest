### 8. Top 100 Sessions by Referring Channel and Medium

This chart shows us which channels and mediums are driving the most traffic (sessions) to our website. It's about getting eyeballs on our products, even if they don't convert immediately.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Re-load df_sessions_marketing_campaigns if not defined
if 'df_sessions_marketing_campaigns' not in locals():
    df_sessions_marketing_campaigns = pd.read_excel('/content/1 year Sessions attributed to marketing campaigns - 2024-12-01 - 2025-11-30.xlsx')

# Re-calculate kpi_sessions_marketing_campaigns if not defined
if 'kpi_sessions_marketing_campaigns' not in locals() or kpi_sessions_marketing_campaigns.empty:
    df_sessions_marketing_campaigns['Referring medium'] = df_sessions_marketing_campaigns['Referring medium'].fillna('Unknown')
    kpi_sessions_marketing_campaigns = df_sessions_marketing_campaigns.groupby(['Referring channel', 'Referring medium'])['Sessions'].sum().reset_index()

kpi_sessions_marketing_campaigns['Referring channel'] = kpi_sessions_marketing_campaigns['Referring channel'].astype(str)
kpi_sessions_marketing_campaigns['Referring medium'] = kpi_sessions_marketing_campaigns['Referring medium'].astype(str)
kpi_sessions_marketing_campaigns['Channel_Medium'] = kpi_sessions_marketing_campaigns['Referring channel'] + ' (' + kpi_sessions_marketing_campaigns['Referring medium'] + ')'
top_20_sessions_marketing = kpi_sessions_marketing_campaigns.nlargest(20, 'Sessions')

plt.figure(figsize=(14, 10)) # Adjusted figure size for 20 bars
sns.barplot(data=top_20_sessions_marketing, x='Sessions', y='Channel_Medium', hue='Channel_Medium', palette='magma', legend=False)
plt.title('Top 20 Sessions by Referring Channel and Medium')
plt.xlabel('Number of Sessions')
plt.ylabel('Referring Channel (Medium)')
plt.tight_layout()
plt.show()
```

#### **Analysis & Actions:**
*   **Finding:** Compare this chart with the 'Total Sales by Referring Channel' chart. High sessions but low sales from a particular channel might mean the traffic isn't high quality, or there's a disconnect between the ad/link and the landing page experience.
*   **Action:** For channels driving high sessions, optimize the landing page experience and calls-to-action to improve conversion. For channels with low sessions, consider adjusting ad creatives or targeting to increase relevant traffic.
